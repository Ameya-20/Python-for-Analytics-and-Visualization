```python
import pandas as pd
import numpy as np
```

```python
#Collecting Records of Division I college basketball seasons of past 10 years
cbb2014 = pd.read_csv('cbb14.csv')
cbb2014
cbb2015 = pd.read_csv('cbb15.csv')
cbb2015
cbb2016 = pd.read_csv('cbb16.csv')
cbb2016
cbb2017 = pd.read_csv('cbb17.csv')
cbb2017
cbb2018 = pd.read_csv('cbb18.csv')
cbb2018
cbb2019 = pd.read_csv('cbb19.csv')
cbb2019
cbb2020 = pd.read_csv('cbb20.csv')
cbb2020
cbb2021 = pd.read_csv('cbb21.csv')
cbb2021
cbb2022 = pd.read_csv('cbb22.csv')
cbb2022
cbb2023 = pd.read_csv('cbb23.csv')
cbb2023
```

```python
#Combining record for past 10 Years
cbb_data = pd.concat([cbb2014, cbb2015, cbb2016, cbb2017, cbb2018, cbb2019, cbb2020, cbb2021, cbb2022, cbb2023])
cbb_data
```

```python
#Number of matches played per year
cbb_data['Year'].value_counts().sort_index()
```

```python
#Champions of every season and their games played and matches won
cbb_data[cbb_data['POSTSEASON'] == 'Champions'][['Year', 'TEAM', 'G', 'W']].sort_values(by = 'Year').reset_index()
```

```python
#Adding win Ratio Column
cbb_data['WIN RATIO'] = cbb_data['W'] * 100 /cbb_data['G']
```

```python
cbb_data[cbb_data['POSTSEASON'] == 'Champions'][['Year', 'TEAM', 'G', 'W', 'WIN RATIO']].sort_values(by = 'Year').reset_index()
```

```python
#Top 8 Teams every season 
cbb_top8 = cbb_data[cbb_data['POSTSEASON'] == 'S16' ][['Year', 'TEAM', 'G', 'W', 'WIN RATIO']].reset_index()
cbb_top8
```

```python
#TEAMS ending top 8 most number of times in past 10 years
cbb_data[cbb_data['POSTSEASON'] == 'S16'][['TEAM']].value_counts().head(10)
```

```python
#Win Ratio of top 8 teams Year wise
cbb_top8[cbb_top8['G'] > 29].groupby('Year')['WIN RATIO'].mean().reset_index()
```

```python
import altair as alt
```

```python
win_ratio_top8 = cbb_top8[cbb_top8['G'] > 29].groupby('Year')['WIN RATIO'].mean().reset_index()
win_ratio_top8
```

```python
alt.Chart(win_ratio_top8).mark_area(
    color="lightblue",
    interpolate='step-after',
    line=True
).encode(
    x='Year',
    y='WIN RATIO'
).transform_filter(alt.datum.symbol == 'GOOG')
```
