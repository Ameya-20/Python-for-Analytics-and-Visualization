```python
import altair as alt
import pandas as pd

# Sample DataFrame
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [5, 3, 2, 4, 1]
})

# Altair chart specification
chart = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
)

# Save the chart to an HTML file
chart.save('chart2.html')
display(chart)

```
